```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import tifffile
```

```python
# Aberration amplitudes introduced by mirror
coeffs = [0.02,0.04,0.06,0.08,0.1,-0.02,-0.04,-0.06,-0.08,-0.1,0]
```

```python
# Extracts the data from the excel filesaved by Zola given the bead identification charecter, folder name,  and number of Zernikes
def extractData(charecter,folder,n):
    filePath =  glob.glob("D:Data/"+folder+"/stacks/Zola("+str(n+3)+")/ZOLA_*"+charecter+".csv")
    #filePath =  glob.glob("D:/Zola(6)Normalized/*.csv")
    files = [open(f) for f in filePath] 
    data = np.zeros((len(files),n))
    for i in range(0,len(files)):
        stringData = [line.rstrip('\n') for line in files[i]][-1].split(",")[-n:]
        data[i] = [np.float(s) for s in stringData]
    return data
```

```python
## Vertical Astigmation
Z = 0 # The Zernike Mode under analysis
nZ = 3 # The total number of Zernike modes fitted used by phase retreival
fig = plt.figure(figsize=(7,7))
key = ['A','B','C','D']

for k,i in enumerate(key):
    data = extractData(i,"2018-08-09-15-04-17-22-Z5Sequential",nZ)
    for c in range(0,nZ):
        plt.subplot(len(key),nZ,(k*nZ)+c+1)
        plt.ylim([-5,5])
        plt.plot(coeffs,data[:,c],'*');
    print(np.polyfit(coeffs,data[:,Z],deg=1))
    
# Prints the slope and intercept for each bead
# Graph: Rows are Beads A,B,C,D; columns are Vertical Astigmation, Defocus, Oblique Astigmation
```
